```python
# Resample the data to monthly intervals -- mean
monthly_gb_mean = df1.resample('M').mean()

# Resample the data to seasonal intervals (3 months) -- mean
seasonal_gb_mean = df1.resample('Q').mean()

# Resample the data to annual intervals -- mean
annual_gb_mean = df1.resample('Y').mean()
```

```python
# annual_gb_mean.reset_index(inplace=True)

# X-axis values (e.g., years)
# years = [2015, 2016, 2017, 2018, 2019]

# List of columns with different units
type_columns = ['CARBON_INTENSITY', 'LOW_CARBON', 'ZERO_CARBON', 'RENEWABLE', 'FOSSIL']

# Create the main plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot data for columns with MWh unit on the left y-axis
for column in type_columns[1:]:
    ax1.plot(annual_gb_mean['DATETIME'], annual_gb_mean[column], label=column, marker='o')

# Set the title and labels for the left y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('MWh', color='b')
ax1.set_title('Energy Generation and Carbon Intensity')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis for the 'CARBON_INTENSITY' column with gCO2/kWh unit
ax2 = ax1.twinx()
ax2.plot(annual_gb_mean['DATETIME'], annual_gb_mean['CARBON_INTENSITY'], label='CARBON_INTENSITY', color='gold', linestyle='dashed', marker='s')
ax2.set_ylabel('gCO2/kWh', color='gold')
ax2.tick_params(axis='y', labelcolor='gold')

# Combine the legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Display the plot
plt.grid(True)
plt.show()

```
